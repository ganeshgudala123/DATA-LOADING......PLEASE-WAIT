# Step 2: Clean → Create SILVER Tables by handling Null values and Duplicates

```python
from pyspark.sql.functions import col, row_number
from pyspark.sql.window import Window

# Load Bronze
bronze_ball = spark.table("ipl_data_analysis.bronze.ball_by_ball")

# Standardize column names
bronze_ball = bronze_ball.toDF(*[c.strip().lower() for c in bronze_ball.columns])

# Remove duplicates
window_spec = Window.partitionBy("match_id", "over_id", "ball_id").orderBy("match_id")

silver_ball = bronze_ball.withColumn("rn", row_number().over(window_spec)) \
    .filter(col("rn") == 1) \
    .drop("rn")

# Handle NULLs (FIXED COLUMN NAMES)
silver_ball = silver_ball.fillna({
    "runs_scored": 0,
    "wides": 0,
    "noballs": 0,
    "extra_runs": 0,
    "bowler_wicket": 0   # ✅ correct column
})

# Fix data types
silver_ball = silver_ball.withColumn("runs_scored", col("runs_scored").cast("int")) \
    .withColumn("wides", col("wides").cast("int")) \
    .withColumn("noballs", col("noballs").cast("int")) \
    .withColumn("extra_runs", col("extra_runs").cast("int")) \
    .withColumn("bowler_wicket", col("bowler_wicket").cast("int"))

# Remove invalid rows
silver_ball = silver_ball.filter(col("match_id").isNotNull())

# Validation
silver_ball = silver_ball.filter((col("runs_scored") >= 0) & (col("runs_scored") <= 6))

# Save to Silver
silver_ball.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ipl_data_analysis.silver.ball_by_ball")
```

```python
from pyspark.sql.functions import col, row_number, sum
from pyspark.sql.window import Window
from pyspark.sql.types import StringType, NumericType

# 1. LOAD FROM BRONZE
df = spark.table("ipl_data_analysis.bronze.ball_by_ball")

# 2. STANDARDIZE COLUMN NAMES
df = df.toDF(*[c.strip().lower() for c in df.columns])

# 3. REMOVE DUPLICATES (only if columns exist)
if all(c in df.columns for c in ["match_id", "over_id", "ball_id"]):
    window_spec = Window.partitionBy("match_id", "over_id", "ball_id").orderBy("match_id")
    df = df.withColumn("rn", row_number().over(window_spec)) \
           .filter(col("rn") == 1) \
           .drop("rn")

# 4. HANDLE NULL VALUES DYNAMICALLY
numeric_cols = [f.name for f in df.schema.fields if isinstance(f.dataType, NumericType)]
string_cols = [f.name for f in df.schema.fields if isinstance(f.dataType, StringType)]

if numeric_cols:
    df = df.fillna(0, subset=numeric_cols)

if string_cols:
    df = df.fillna("Unknown", subset=string_cols)

# 5. CAST IMPORTANT COLUMNS IF PRESENT
for c in ["runs_scored", "wides", "noballs", "extra_runs", "bowler_wicket"]:
    if c in df.columns:
        df = df.withColumn(c, col(c).cast("int"))

# 6. REMOVE INVALID ROWS
if "match_id" in df.columns:
    df = df.filter(col("match_id").isNotNull())

# 7. OPTIONAL VALIDATION (only if column exists)
if "runs_scored" in df.columns:
    df = df.filter((col("runs_scored") >= 0) & (col("runs_scored") <= 6))

# 8. FINAL NULL CHECK (should be 0)
df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in df.columns
]).show()

# 9. SAVE TO SILVER (DELTA)
df.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ipl_data_analysis.silver.ball_by_ball")
```

# STEP-3 GOLD LAYER

# 1. Match Winning Analysis

```python
from pyspark.sql.functions import col, when

# Load from Silver layer
silver_match = spark.table("ipl_data_analysis.silver.match")

# Add toss_win_match_win column
gold_df = silver_match.withColumn(
    "toss_win_match_win",
    when(col("toss_winner") == col("match_winner"), "Yes")
    .otherwise("No")
)

# Save Gold table
gold_df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.match_analysis")
```

# Top performing Players

```python
from pyspark.sql.functions import sum, col, count, round

# Load from Silver
ball_by_ball_df = spark.table("ipl_data_analysis.silver.ball_by_ball")

# Top run scorers
top_runs = ball_by_ball_df.groupBy("striker") \
    .agg(sum("runs_scored").alias("total_runs")) \
    .orderBy(col("total_runs").desc())

# Save to Gold
top_runs.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.top_run_scorers")

# Top Wicket Takers
top_wickets = ball_by_ball_df.filter(col("out_type").isNotNull()) \
    .groupBy("bowler") \
    .agg(count("out_type").alias("total_wickets")) \
    .orderBy(col("total_wickets").desc())

top_wickets.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.top_wicket_takers")

# Strike Rate Analysis
strike_rate = ball_by_ball_df.groupBy("striker") \
    .agg(
        sum("runs_scored").alias("runs"),
        count("*").alias("balls")
    ).filter(col("balls") >= 300) \
    .withColumn(
        "strike_rate",
        round((col("runs") / col("balls")) * 100, 2)
    ).orderBy(col("strike_rate").desc())

strike_rate.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.strike_rate_analysis")
```

# Toss Impact Analysis

```python
from pyspark.sql.functions import col, when

toss_impact_individual_matches = spark.table("ipl_data_analysis.silver.match") \
    .select(
        col("match_id"),
        col("toss_winner"),
        col("toss_name").alias("toss_decision"),
        col("match_winner")
    ).withColumn(
        "match_outcome",
        when(col("toss_winner") == col("match_winner"), "Won")
        .otherwise("Lost")
    ).orderBy("match_id")

# Save as Gold table
toss_impact_individual_matches.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("ipl_data_analysis.gold.toss_impact_individual_matches")
```

# Venue High Scoring Analysis

```python
from pyspark.sql.functions import sum, col

# Load from Silver
ball_by_ball_df = spark.table("ipl_data_analysis.silver.ball_by_ball")
match_df = spark.table("ipl_data_analysis.silver.match")

match_runs = ball_by_ball_df.groupBy("match_id") \
    .agg(sum("runs_scored").alias("match_runs"))

venue_analysis = match_runs.join(match_df, match_runs.match_id == match_df.match_id) \
    .groupBy("venue_name") \
    .agg(sum("match_runs").alias("total_runs")) \
    .orderBy(col("total_runs").desc())

venue_analysis.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.venue_analysis")
```

# Player Impact Score

```python
from pyspark.sql.functions import sum, count, col, round

# Load from Silver
ball_by_ball_df = spark.table("ipl_data_analysis.silver.ball_by_ball")

impact_score = ball_by_ball_df.groupBy("striker") \
    .agg(
        sum("runs_scored").alias("runs"),
        count("*").alias("balls")
    ).withColumn(
        "impact_score",
        round(col("runs") + (col("runs") / col("balls") * 100), 2)
    ).orderBy(col("impact_score").desc())

impact_score.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("ipl_data_analysis.gold.player_impact_score")
```

# TOP SCORING BATSMEN PER SEASON 

```python
top_scoring_batsmen_per_season = spark.sql("""
SELECT player_name, season, total_runs
FROM (
    SELECT 
        p.player_name,
        b.season,
        SUM(b.runs_scored) AS total_runs,
        DENSE_RANK() OVER (PARTITION BY b.season ORDER BY SUM(b.runs_scored) DESC) AS rnk
    FROM ipl_data_analysis.silver.ball_by_ball b
    JOIN ipl_data_analysis.silver.match m 
        ON b.match_id = m.match_id
    JOIN ipl_data_analysis.silver.player p 
        ON b.striker = p.player_id
    GROUP BY p.player_name, b.season
)
WHERE rnk = 1
ORDER BY season
""")

# Save as Gold table
top_scoring_batsmen_per_season.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("ipl_data_analysis.gold.top_batsman_per_season")

# Display
display(top_scoring_batsmen_per_season)
```
